```python
import matplotlib.pyplot as plt
import numpy as np

voltage = np.array(range(157))
frequency = np.sqrt(voltage/(4*(np.pi**2)*(5*10**-9)*(1.6*10**-19)))
plt.xlabel("Voltage(v)")
plt.ylabel("Frequency(Hz)")
plt.title("Graph of frequency against voltage")
plt.plot(voltage,frequency)
plt.grid(b=None, which='major', axis='both')
```
